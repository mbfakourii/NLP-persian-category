```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize.treebank import TreebankWordDetokenizer
from tqdm import tqdm
import re

nltk.download('punkt')
```

```python
train_digimag = pd.read_csv("/content/drive/MyDrive/Deep/NLP/digimag/train.csv", sep='	' )
test_digimag = pd.read_csv("/content/drive/MyDrive/Deep/NLP/digimag/test.csv", sep='	')
dev_digimag = pd.read_csv("/content/drive/MyDrive/Deep/NLP/digimag/dev.csv", sep='	')


train_persian_news = pd.read_csv("/content/drive/MyDrive/Deep/NLP/persian_news/train.csv", sep='	' )
test_persian_news = pd.read_csv("/content/drive/MyDrive/Deep/NLP/persian_news/test.csv", sep='	')
dev_persian_news = pd.read_csv("/content/drive/MyDrive/Deep/NLP/persian_news/dev.csv", sep='	')


stopwords = open('/content/drive/MyDrive/Deep/NLP/persian_stop_words.txt', 'r').read().split('\n')
```

```python
def pre_processing(data , saveFileName):
  # get all words in stop words file
  for i in tqdm(range(len(data['content']))):
    word_tokens = word_tokenize(data['content'][i])

    filtered_sentence = []

    for w in word_tokens:
        if w not in stopwords:
            filtered_sentence.append(w)
    text = TreebankWordDetokenizer().detokenize(filtered_sentence).replace('\\', '')

    text = text.replace('\u200c', ' ')

    """ Remove all html tags """
    cleaner = re.compile('<.*?>')
    text = re.sub(cleaner, '', text)

    #remove usernam
    text = re.sub(r"\@[\w.-_]+", " ", text)

    # remove hashtag
    text = text.replace('#', ' ')
    text = text.replace('_', ' ')

    # removing extra spaces, hashtags
    text = re.sub("#", "", text)


    # removing weird patterns
    weird_pattern = re.compile(
      "["
      u"\U0001F600-\U0001F64F"  # emoticons
      u"\U0001F300-\U0001F5FF"  # symbols & pictographs
      u"\U0001F680-\U0001F6FF"  # transport & map symbols
      u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
      u"\U00002702-\U000027B0"
      u"\U000024C2-\U0001F251"
      u"\U0001f926-\U0001f937"
      u'\U00010000-\U0010ffff'
      u"\u200d"
      u"\u2640-\u2642"
      u"\u2600-\u2B55"
      u"\u23cf"
      u"\u23e9"
      u"\u231a"
      u"\u3030"
      u"\ufe0f"
      u"\u2069"
      u"\u2066"
      u"\u2013"
      u"\u2068"
      u"\u2067"
      "]+", flags=re.UNICODE)

    text = weird_pattern.sub(r'', text)

    data['content'][i] = text

  # save to csv file
  data.to_csv(saveFileName, sep='\t', encoding='utf-8' , index=False)
```

```python
pre_processing(train_digimag,"/content/drive/MyDrive/Deep/NLP/digimag/fixed_train.csv")
pre_processing(test_digimag,"/content/drive/MyDrive/Deep/NLP/digimag/fixed_test.csv")
pre_processing(dev_digimag,"/content/drive/MyDrive/Deep/NLP/digimag/fixed_dev.csv")


pre_processing(train_persian_news,"/content/drive/MyDrive/Deep/NLP/persian_news/fixed_train.csv")
pre_processing(test_persian_news,"/content/drive/MyDrive/Deep/NLP/persian_news/fixed_test.csv")
pre_processing(dev_persian_news,"/content/drive/MyDrive/Deep/NLP/persian_news/fixed_dev.csv")
```
